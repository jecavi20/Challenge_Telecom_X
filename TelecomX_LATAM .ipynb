```python
import pandas as pd
import matplotlib.pyplot as plt

archivo = 'https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json'

df = pd.read_json(archivo)

df.head(10)
```

#ðŸ”§ TransformaciÃ³n

```python
datos_customer= pd.json_normalize(df['customer'])

datos_customer
```

```python
datos_phone= pd.json_normalize(df['phone'])

datos_phone.head(10)
```

```python
datos_internet = pd.json_normalize(df['internet'])
datos_internet.head(10)
```

```python
datos_account = pd.json_normalize(df['account'])
datos_account
```

```python
df.drop(['customer','phone','internet','account'], axis=1, inplace=True)
```

```python
df
```

```python
datos = df.join([datos_customer,datos_phone,datos_internet,datos_account])
```

```python
datos.head(10)
```

```python
datos.columns
```

```python
datos.info()
```

```python
datos['Charges.Total'].sample(10)
```

```python
datos['Charges.Total'] = datos['Charges.Total'].str.replace(' ', '0').replace(',','.')
```

```python
datos['Charges.Total'] = datos['Charges.Total'].astype(float)
```

```python
datos.info()
```

```python
datos.columns
```

```python
datos.columns = ['ID_cliente', 'Churn', 'Genero', 'Tercera_edad', 'Partner',
'Dependientes', 'Antiguedad', 'Servicio_telefonico', 'Lineas_multiples',
'Servicio_de_internet', 'Seguridad_en_linea', 'Copia_de_seguridad_en_linea', 'Proteccion_del_dispositivo',
'Soporte_tecnico', 'Transmision_de_TV', 'Transmision_de_peliculas', 'Contrato',
'Facturacion_electronica', 'Metodo_de_pago', 'Cargos_mensuales',
'Cargos_totales']
```

```python
datos.head(10)
```

```python
datos['Churn'].unique()
```

```python
datos.query('Churn == ""')
```

```python
churn_vacios = datos[datos['Churn'] == ''].index
datos = datos.drop(churn_vacios)
```

```python
datos.query('Churn == ""')
```

```python
datos['Cuentas_Diarias'] = datos['Cargos_mensuales'] / 30
```

```python
#Grafico de comparacion de los clientes que abandonaron el servicio y los que no

colores = ['deepskyblue', 'lightcoral','springgreen','darkviolet']
fig, axs = plt.subplots(1, 1, figsize=(8, 5))
plt.subplots_adjust(wspace=0.2, hspace=0.4)
datos['Churn'].value_counts().plot(ax=axs, kind='bar', color=colores)
axs.set_title('ComparaciÃ³n de evasiÃ³n de clientes (Churn)')
axs.set_xlabel('')
axs.set_ylabel('Cantidad')
axs.set(xticklabels=['Clientes Activos', 'Clientes Retirados'])
axs.set_ylim(0,6000)
axs.spines['top'].set_visible(False)
axs.spines['right'].set_visible(False)
axs.set_xticklabels(axs.get_xticklabels(), rotation=360, ha='center')
porcentaje_churn = (datos['Churn'].value_counts()/datos['Churn'].value_counts().sum()) * 100
churn_p = {'Pno':round(porcentaje_churn.iloc[0],1),'Pyes':round(porcentaje_churn.iloc[1],1)}
churn_p = pd.Series(churn_p)
for i, count in enumerate(datos['Churn'].value_counts()):
    axs.text(i, count+100, str(f'{count} ({churn_p.iloc[i]}%)'), ha='center', fontsize=12)

plt.show()
```
